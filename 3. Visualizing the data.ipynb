```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Cars = pd.read_csv('Cars_dataset.csv')
Cars.head()
```

```python
Cars.describe()
```

```python
# Calculating the countinuous variable(min, max)
# Calculating the Central Tendency (mean, median, mode)
Cars.describe()
```

```python
Cars['min_price_inLakhs'].mode()[0]
```

```python
# Measures of Dispersion(standard deviation, variance)
Cars['min_price_inLakhs'].var()
```

```python
Cars['min_price_inLakhs'].std()
```

# Uni-variate Analysis

```python
plt.figure(figsize = (4,3))
sns.countplot(x='Transmission', data=Cars , color = 'darkblue')
plt.title('Count of Cars with Transmission')
plt.show()
```

```python
plt.figure(figsize = (4,3))
sns.kdeplot(data = Cars, x = Cars['min_price_inLakhs'], color = 'red')
plt.title('Flow of Minimum Car Prices ')
plt.show()
```

```python
plt.figure(figsize = (4,3))
plt.hist(Cars['min_price_inLakhs'],color = 'pink')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
```

#  Bi-variate Analysis

```python
Cars.groupby(Cars['Transmission'])['min_price_inLakhs'].mean() # Grouping the Cars by transmission and finding mean of price
```

```python
Cars[['min_price_inLakhs','max_price_inLakhs']].corr() 
```

```python
Cars[['min_price_inLakhs','max_price_inLakhs']].corr(method = 'spearman')
```

```python
pd.pivot_table(Cars,values='min_price_inLakhs',index='Transmission')
```

```python
plt.figure(figsize = (4,3))
sns.violinplot(y ='Transmission', x ='min_price_inLakhs', data = Cars, color = 'darkgreen')
plt.title('Price Distribution by Transmission Type')
plt.xlabel('Minimum Price (in Lakh)')
plt.ylabel('Transmission')
plt.show()
```

```python
plt.figure(figsize=(9,3))
plt.bar(Cars['Car Name'], Cars['max_price_inLakhs'])
plt.xticks(rotation=45, ha='right')
plt.xlabel('Car Model')
plt.ylabel('Maximum Price (in Lakhs)')
plt.title('Maximum Price by Car Model')
plt.show()
```
